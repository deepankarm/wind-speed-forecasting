```python
import os, time
import pandas as pd

from selenium.webdriver import Firefox
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import *

from selenium.webdriver.support.ui import Select
```

```python
CORE_PATH = "/Users/deepankar/my_own/code_n_stuff/playground/aliva"
DRIVER_PATH = "/Users/deepankar/my_own/code_n_stuff/playground/aliva/geckodriver"
```

```python
%%time

list_of_dates = []

for year in range(2009, 2021):
    for month in range(1, 13):
        if year == 2009 and month in range(1, 9):
            continue
        if year == 2020 and month in range(2, 13):
            continue
        print(f"\n\nProcessing for year: `{year}`, month: `{month}`")
        CURRENT_URL = f"https://www.timeanddate.com/weather/india/surat/historic?month={month}&year={year}"
        
        os.system('killall -9 firefox-bin')

        driver = Firefox(executable_path=DRIVER_PATH)
        driver.get(CURRENT_URL)

        try:
            element = WebDriverWait(driver, 30).until(
                expected_conditions.presence_of_element_located((By.XPATH, "//*[@id='wt-his-select']"))
            )
        except NoSuchElementException as e:
            print("ERROR: Web page didn't open properly")
            raise KeyboardInterrupt

        dropdown_box = driver.find_element_by_xpath("//*[@id='wt-his-select']")
        select = Select(dropdown_box)

        options = [x for x in dropdown_box.find_elements_by_tag_name("option")]

        for date_value_element in options:
            date_text = date_value_element.text
            print(f"Current Date: {date_text}")
            date_value = date_value_element.get_attribute("value")
            select.select_by_value(date_value)
            
            time.sleep(1)
            try:
                table_id = driver.find_element_by_xpath("//*[@id='wt-his']/tbody")
                rows = table_id.find_elements(By.TAG_NAME, "tr")

                for row_index, row in enumerate(rows, start=1):

                    clock_time = row.find_elements(By.TAG_NAME, "th")[0].text.split("\n")[0]
                    temperature = row.find_elements(By.TAG_NAME, "td")[1].text
                    weather = row.find_elements(By.TAG_NAME, "td")[2].text
                    wind_speed = row.find_elements(By.TAG_NAME, "td")[3].text
                    humidity = row.find_elements(By.TAG_NAME, "td")[5].text
                    barometer = row.find_elements(By.TAG_NAME, "td")[6].text
                    visibility = row.find_elements(By.TAG_NAME, "td")[7].text

                    list_of_dates.append({ 
                        date_text : 
                            {
                                'Clock_Time' : clock_time,
                                'Temperature' : temperature,
                                'Weather' : weather,
                                'Wind_Speed' : wind_speed,
                                'Humidity' : humidity,
                                'Visibility' : visibility
                            }
                    })
            except (StaleElementReferenceException, NoSuchElementException) as e:
                print(e)
                continue
            
```

```python
%%time
dataset = pd.concat([
    pd.DataFrame.from_dict(dates_dict, orient='index') 
    for dates_dict in list_of_dates])
```
